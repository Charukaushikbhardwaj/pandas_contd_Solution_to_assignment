```python
import pandas as pd
import numpy as np
from pandas import Series,DataFrame

url='https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv'
data=pd.read_csv(url)
#df=pd.DataFrame(data)
#df.head()
data.head()
```

```python
# Question 1-Delete unnamed columns.
Data=data.iloc[0:,1:7]
Data.head(10)
```

```python
# Question 2- Show the distribution of male and female
data["Gender"].value_counts()
```

```python
# Question 3-Show the top 5 most preferred names, cell 10-14

Name_Count=Data[['Name','Count']]
Name_Count.head()
```

```python
NC=Name_Count.groupby(['Name']).sum()
NC.head()
```

```python
NC.sort_values("Count",ascending=0).head()
```

```python
# Qustion 4-What is the median name occurence in the dataset
NC[NC.Count==NC.Count.median()]
```
